# Automatic Diagram chassing in double chain complex


The translation of your diagram in terms of a double chain complex is not part of this file. The coordinates of the vertex on a grid $\mathbb{Z}^2$ has to be done before using it.

The places to modify something are explicitely mentioned.

Make sure to exectute the function $start()$ before starting a new computation

There is by default the example of the snake lemma.

```python
start()
```

## GIVE THE DIAGRAM
### 1 GIVE THE VERTICES

In this first step you just have to declare the vertex using `add_vertex` of your diagram, with a label for each one. For the following point, it is important that the labeling is injective.

If the vertex must be a zero object precise it by seting the parameter $is\_zero$ to True (in particular the program does not care if the vertex is named "ZERO")

There is no need to add kernels or cokernels at the edge of the diagram, it will be done automaticaly later.


```python
#add_vertex(i,j,"label",is_zero=True)

#the vertex of the snake lemma:
#MODIFY THERE:
add_vertex(0,0,"X1")
add_vertex(0,1,"X2")
add_vertex(0,2,"X3")
add_vertex(1,0,"Y1")
add_vertex(1,1,"Y2")
add_vertex(1,2,"Y3")
```

From this point no vertex should be added, otherwise the result is going to be wrong.

The function connect is there to declare the edges between the declared vertex, and to extend (as soon as possible) the double chain complex by adding at the edge the kernels and cokernels (it will be useful for computation)

```python
connect()
```

Use the function `add_mono`, `add_epi` and `add_iso` (which take the same argument) to declare what edges (between two objects you declared) are monomorphism, epimorphism or isomorphism

For this step it is important that the labeling is injective.

## EXTRACT PROOF
### 1 IF YOU KNOW WHAT YOU WANT:
Translate your statement using the following conventions and just look at the proof:

`('is_zero_obj',name)`: the object name is $0$

`('is_mono',start,end)`: the map $start \to end$ is a mono

`('is_epi',start,end)`: the map $start \to end$ is an epi

`('is_iso',start,end)`: the map $start \to end$ is an iso

`('is_zero',start,end)`: the map $start \to end$ is the zero map

`('is_exact',start,mid,end)`: the composition of $start \to mid$ and $mid \to end$ is exact

`('is_exact(connected)',A,B,C,D)`: if $A\to B\to 0$ and $0\to C\to D$ are part of the diagram there is a conection morphism $\delta: B\to C$ such that the sequence $A\to B \to C \to D$

## Other examples
### 1 four lemma, injective

```python
# A->B->C->D
# |  |  |  |
# E->F->G->H
# If the rows are exact, A->E is epi, B->F  and D->H are mono then C->G is mono

start()

add_vertex(0,0,"A")
add_vertex(0,1,"B")
add_vertex(0,2,"C")
add_vertex(0,3,"D")
add_vertex(1,0,"E")
add_vertex(1,1,"F")
add_vertex(1,2,"G")
add_vertex(1,3,"H")

connect()

add_epi("A","E")
add_mono("B","F")
add_mono("D","H")


add_exact_hline("B","C")
add_exact_hline("E","F")

statement=('is_mono',"C","G")

#display_proof(statement)
```

### 2 Five lemma

```python
#A1->A2->A3->A4->A5
#|   |   |   |   |
#B1->B2->B3->B4->B5

#If the two rows are exact, A2->B2 and A4->B4 are mono, A1->B1 is epi then A3->B3 is mono

start()

add_vertex(0,0,"A1")
add_vertex(0,1,"A2")
add_vertex(0,2,"A3")
add_vertex(0,3,"A4")
add_vertex(0,4,"A5")
add_vertex(1,0,"B1")
add_vertex(1,1,"B2")
add_vertex(1,2,"B3")
add_vertex(1,3,"B4")
add_vertex(1,4,"B5")

connect()

add_iso("A2","B2")
add_iso("A4","B4")
add_epi("A1","B1")
add_mono("A5","B5")

add_exact_hline("A2","A4")
add_exact_hline("B2","B4")

statement=('is_iso',"A3","B3")

#display_proof(statement)
```

### 3 the $3\times 3$ lemma

```python
#     0    0    0
#     |    |    |
# 0->A_11->A_12->A_13
#     |     |     |
# 0->A_21->A_22->A_23
#     |     |     |
# 0->A_31->A_32->A_13

#If all the rows but the first one and all the columns are exact the the first line is exact


start()

add_vertex(0,0,"A_11")
add_vertex(0,1,"A_12")
add_vertex(0,2,"A_13")
add_vertex(1,0,"A_21")
add_vertex(1,1,"A_22")
add_vertex(1,2,"A_23")
add_vertex(2,0,"A_31")
add_vertex(2,1,"A_32")
add_vertex(2,2,"A_33")

add_vertex(-1,0,"ZERO_01",is_zero=True)
add_vertex(-1,1,"ZERO_02",is_zero=True)
add_vertex(-1,2,"ZERO_03",is_zero=True)
add_vertex(0,-1,"ZERO_10",is_zero=True)
add_vertex(1,-1,"ZERO_20",is_zero=True)
add_vertex(2,-1,"ZERO_30",is_zero=True)

connect()

add_exact_hline("ZERO_20","A_22")
add_exact_hline("ZERO_30","A_32")

add_exact_vline("ZERO_01","A_21")
add_exact_vline("ZERO_02","A_22")
add_exact_vline("ZERO_03","A_23")

statement1=('is_exact',"ZERO_10","A_11","A_12")
statement2=('is_exact',"A_11","A_12","A_13")

#display_proof(statement1)
#display_proof(statement2)
```
