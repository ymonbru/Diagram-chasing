In this fisrt step you just have to declare the vertex using `add_vertex` of your diagram, with a label for each one. For the following point, it is important that the labeling is injective.

If the vertex must be a zero object precise it by seting the parameter $is\_zero$ to True (in particular the program does not care if the vertex is named "ZERO")

There is no need to add kernels or cokernels at the edge of the diagram, it will be donne automaticaly later.


```python
#add_vertex(i,j,"label",is_zero=True)

#the vertex of the snake lemma:
#MODIFY THERE:
add_vertex(0,0,"X1")
add_vertex(0,1,"X2")
add_vertex(0,2,"X3")
add_vertex(1,0,"Y1")
add_vertex(1,1,"Y2")
add_vertex(1,2,"Y3")
```

From this point no vertex should be added, otherwise the result is going to be wrong.

The function connect is there to declare the edges between the declared vertex, and to extend (as soon as possible) the double chain complex by addingg at the edge the kernels and cokernels (it will be usefull for computation)

```python
connect()
```

### 1 IF YOU KNOW WHAT YOU WANT:

`('is_zero_obj',name)`: the object name is $0$

`('is_mono',start,end)`: the map $start \to end$ is a mono

`('is_epi',start,end)`: the map $start \to end$ is an epi

`('is_iso',start,end)`: the map $start \to end$ is an iso

`('is_zero',start,end)`: the map $start \to end$ is the zero map
